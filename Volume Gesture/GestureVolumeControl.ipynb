```python
import sys
sys.path.insert(1,r"C:\Users\ayush\OneDrive\Desktop\handTracking\HandTrackingModule.ipynb")
import cv2
import mediapipe as mp
import time
import ipynb.fs.full.HandTrackingModule as htm
import math
from ctypes import cast, POINTER
from comtypes import CLSCTX_ALL
from pycaw.pycaw import AudioUtilities, IAudioEndpointVolume
import numpy as np
```

```python
detector=htm.handDetector(detectionCon=0.7)
cap=cv2.VideoCapture(0)
cTime=0
pTime=0
wCam=1280
hCam=1280
cap.set(3,wCam)
cap.set(4,hCam)


devices = AudioUtilities.GetSpeakers()
interface = devices.Activate(
    IAudioEndpointVolume._iid_, CLSCTX_ALL, None)
volume = cast(interface, POINTER(IAudioEndpointVolume))
#volume.GetMute()
#volume.GetMasterVolumeLevel()
# (-65.25, 0.0, 0.03125)
# range of volume is from -65.25 to 0.0
volRange=volume.GetVolumeRange() # return the volume range from (-65,25) to (0)
minVol=volRange[0]
maxVol=volRange[1]
volume.SetMasterVolumeLevel(0, None)

while True:
    success,img=cap.read()
    img=detector.findHands(img)
    lmList=detector.findPosition(img,draw=False)
    cv2.rectangle(img,(30,150),(85,500),(0,255,0),3)
    if len(lmList)!=0:
        x1,y1=lmList[4][1],lmList[4][2]
        x2,y2=lmList[8][1],lmList[8][2]
        cx,cy=((x1+x2)//2),((y1+y2)//2)
        cv2.line(img,(x1,y1),(x2,y2),(255,0,255),5)
        cv2.circle(img,(x1,y1),10,(255,0,255),cv2.FILLED)
        cv2.circle(img,(x2,y2),10,(255,0,255),cv2.FILLED)
        cv2.circle(img,(cx,cy),10,(255,0,255),cv2.FILLED)
        length=math.hypot(x2-x1,y2-y1)
        print(length)
        
        # hand range from 30-300
        # volume.GetVolumeRange() range from (-65,25) to (0)
        # np.interp is use to convert a particular variable of range to other range
        vol=np.interp(length,[50,300],[minVol,maxVol])
        v=np.interp(length,[50,300],[500,150])
        cv2.rectangle(img,(30,int(v)),(85,500),(0,255,0),cv2.FILLED)
        volume.SetMasterVolumeLevel(vol, None) 
        
        volper=np.interp(length,[50,300],[0,100])
        cv2.putText(img,f'{int(volper)}%',(30,140),cv2.FONT_HERSHEY_PLAIN,3,(0,250,0),3)
        
       # print(lmList[4],lmList[8])
       # print(lmList)
    # fps
    cTime=time.time()
    fps=1/(cTime-pTime)
    pTime=cTime 
    cv2.putText(img,"FPS:"+str(int(fps)),(20,70),cv2.FONT_HERSHEY_PLAIN,5,(0,250,0),5)
    cv2.imshow('img',img)
    
    if cv2.waitKey(1) & 0xFF==ord('a'):
        break
cap.release()
cv2.destroyAllWindows()
    
```
